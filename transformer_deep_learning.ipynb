**tokenizaion**

```python
eng_tokenizer = Tokenizer()
eng_tokenizer.fit_on_texts(english)

poshto_tokenizer = Tokenizer()
poshto_tokenizer.fit_on_texts(poshto)


print("English Tokenizer:")
print(eng_tokenizer.word_index)
print(eng_tokenizer.word_counts)
print(len(eng_tokenizer.word_index) + 1)


#====================

print("\nPoshto Tokenizer:")
print(poshto_tokenizer.word_index)
print(poshto_tokenizer.word_counts)
print(len(poshto_tokenizer.word_index) + 1)
```

**Convert Text → Sequences + Padding**

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
eng_sequence = eng_tokenizer.texts_to_sequences(english)
poshto_sequence = poshto_tokenizer.texts_to_sequences(poshto)

eng_padded = pad_sequences(eng_sequence, padding="post")
poshto_padded = pad_sequences(poshto_sequence, padding='post')
```

**Embedding Layer**


```python
from tensorflow.keras.layers import Embedding
print(len(eng_tokenizer.word_index) + 1)
Embedding_layer = Embedding(input_dim = len(eng_tokenizer.word_index)+1 ,output_dim=64)

embedded_output = Embedding_layer(eng_padded)

print(f'embedding shape :\n {embedded_output.shape}')
print(f'first embeddedd sentence :\n {embedded_output[0]}')
```
